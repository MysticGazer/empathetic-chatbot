```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# pastikan dulu kamu punya ini
nltk.download('stopwords')
nltk.download('wordnet')

# load data hasil eksplorasi
df = pd.read_csv("data/processed/emotion_dataset_clean.csv")
df.head()
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)  # hapus URL
    text = re.sub(r'\@w+|\#','', text)  # hapus mention dan hashtag
    text = text.lower()  # lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # hapus tanda baca
    text = re.sub(r'\d+', '', text)  # hapus angka
    tokens = text.split()  # tokenisasi sederhana
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]  # hapus stopwords dan lemmatize
    return " ".join(tokens)

```

```python
df['clean_text'] = df['text'].apply(clean_text)
df[['text', 'clean_text', 'emotion_label']].head(10)
```

```python
df['text_len_before'] = df['text'].apply(lambda x: len(x.split()))
df['text_len_after'] = df['clean_text'].apply(lambda x: len(x.split()))

df[['text_len_before', 'text_len_after']].describe()
```

```python
df.to_csv("data/processed/emotion_dataset_preprocessed.csv", index=False)
print("✅ Data preprocessing selesai dan disimpan ke /data/processed/")
```
